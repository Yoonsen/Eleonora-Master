```python
corpus = pd.read_excel('riverton.xlsx')
```

```python
corpus.head(5)
```

```python
def gender_from_authors(corp):
    genders = dict()
    for w in corp.authors.apply(lambda x: x.strip().split(',')[-1]):
        genders[w.strip()] = d.get_gender(w.strip().split(' ')[0])
    return genders
```

```python
gender_from_authors(corpus)
```

# Her kombineres original med dh-versjon

```python
combo = pd.concat([ocorpus[ocorpus.urn.str.startswith('URN')].set_index('urn'), corpus.set_index('urn')], axis = 1)
```

```python
test = combo[["title" ,"authors", 'Forfatters kjønn (K/M)']].copy()
```

#### Legger gender i en egen kolonne

```python
test['gender'] = test.authors.apply(lambda x: d.get_gender(x.split(',')[-1].strip().split(' ')[0]))
```

```python
test
```

```python
test['egender'] = test['Forfatters kjønn (K/M)'].apply(lambda x: 'male' if x == 'M' else 'female' if x == 'K' else 'X')
```

Tester diskrepans i kjønnskategorisering

```python
test[test.egender != test.gender]
```
