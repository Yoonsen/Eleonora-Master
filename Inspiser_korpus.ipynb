```python
import pandas as pd
import dhlab as dh
import re
```

```python
korpus = pd.read_excel("Korpus.xlsx").fillna('')
```

```python
# black magic aka regulære uttrykk og pandas gymnastikk
# Faller inn under aktiviteten "data cleaning"

korpus['urn'] = korpus.URN[korpus.URN.str.startswith('http')].apply(lambda x: x.split('/')[-1])
```

```python
corpus = korpus.dropna()
```

```python
corpus.sample(4)
```

## Kollokasjon

Henter ut ordene rundt søkeordet

Sjekk parametrene, og juster og eksperimenter

```python
# data for sammenligning

tot = dh.totals(300000)
```

# lag som prosent

```python
tot_rel = tot.freq*100/tot.freq.sum()
```

```python
tot_rel.head()
```

## Hent ut kollokasjon

```python
coll = dh.Collocations(corpus=corpus, words = ['liket','lik','Liket'], before = 12, after = 12).frame
```

```python
coll = coll.sort_values(by="counts", ascending=False)
```

```python
coll.head()
```

```python
coll_rel = coll['counts']*100/coll['counts'].sum()
```

# Ser på forskjellen mellom kollokasjonen og referansen

```python
diff = coll_rel/tot_rel
```

```python
diff.sort_values(ascending=False).head(10)
```

## Sjekk frekvenser av enkeltord

```python
coll.loc["Gunnarstranda"]
```

### Lager en kombinasjonsramme

```python
combo = pd.concat([coll.counts, diff], axis = 1)
```

```python
combo.columns = ['counts', 'diff']
```

```python
combo
```

```python
combo.loc[['han','hun']]
```

```python
combo[combo['diff'].isna()].sort_values(by='counts', ascending = False).head(5)
```
