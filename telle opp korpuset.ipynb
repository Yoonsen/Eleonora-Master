```python
resultat=list()
for x in words:
    for key in words [x]["former"]:
        forms=words [x]["former"][key].split()
        if len(forms)==1:
            word=forms[0]
        else:
            word=forms[1]
        resultat.append((x, word))
word_df=pd.DataFrame(resultat, columns=["grunnform", "bøyd form"])
word_df
```
